# Insert CSV into MongoDB with PyMongo

```python
#This document is to setup the MongoDB database to store our csv file
```

```python
import pandas as pd
import pymongo
import json
```

##### Create Mongo client variable to point to the location on our computer

```python
client = pymongo.MongoClient("mongodb://localhost:27017")
```

##### Read in data

```python
df = pd.read_csv('Project/Google_App_Store_Data_Clean.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

##### Convert DF to dictionary (json)

```python
data = df.to_dict(orient = "records")
```

```python
data
```

##### Create DB with the name below

```python
db = client["Google_App_Store_Data"]
```

##### Create a collection with name "Google_App_Store_Data"

```python
db.Google_App_Store_Data.insert_many(data)
```
