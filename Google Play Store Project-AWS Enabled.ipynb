```python
#Removing columns that are unneccessary
sample_df.drop(['App Id','Installs','Free','Minimum Installs','Currency','Developer Website','Developer Email','Editors Choice','Scraped Time','Content Rating','Last Updated','Developer Id','Minimum Android','Privacy Policy'], axis=1, inplace=True)
```

```python
sample_df.head(5)
```

```python
sample_df.count()
```

```python
#Removing columns that are null 
df = sample_df.dropna(how='any')
```

```python
df.drop_duplicates()
```

```python
#Deleting data with Maximum installs less than 100
df = df[df['Maximum Installs'] > 100] 
```

```python
#Deleting data with Rating count less than 50
df = df[df['Rating Count'] > 50] 
```

```python
df.count()
```

```python
#Converting InAppPurchases and Ad Supported from string to binary values 
df['In App Purchases'].dtype
```

```python
df.head()
```

```python
df.info()
```

```python
df["In App Purchases"] = df["In App Purchases"].astype(int)
```

```python
df['In App Purchases'].dtype
```

```python
df["Ad Supported"] = df["Ad Supported"].astype(int)
```

```python
df['Ad Supported'].dtype
```

```python
for col in df:
  print(df[col].unique())
```

```python
df.head()
```

```python
#Released column converted to Date
df['Released'] = pd.to_datetime(df['Released'])
```

```python
df.info()
```

```python
#Chagning Column Name
df = df.rename(columns={"Size": "Size (Mb)"})
#Replacing M with nothing
df["Size (Mb)"] = df["Size (Mb)"].str.replace('M', '', regex=True)


df.head()
```
