```python
import pandas as pd
from path import Path
import numpy as np
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

file_path = Path("Project/Google_App_Store_Data_Clean.csv")
app_data = pd.read_csv(file_path, index_col = 1)
app_data.head()
```

```python
app_data = app_data.drop(["Unnamed: 0","Month_Name"],axis=1)
app_data.head()
```

```python
app_data['Rating'].median()
```

```python
app_data['Maximum Installs'].median()
```

```python
app_data['Good_App'] = np.where((app_data['Rating']>=4.2) & (app_data['Rating Count']>=250) & (app_data['Maximum Installs']>=43901.5),1,0)
app_data.head()
```

```python
app_data_encoded = pd.get_dummies(app_data, columns=['Category'])
app_data_encoded.head()
```

```python
app_data_encoded.columns
```

```python
app_data_encoded = app_data_encoded.drop(['Released','Size (Mb)'],axis=1)
app_data_encoded.head()
```

```python
app_data_encoded.dtypes
```

```python
data_scaler = StandardScaler()
```

```python
app_data_scaled = data_scaler.fit_transform(app_data_encoded)
```

```python
X = app_data_encoded.copy()
X = X.drop(["Rating","Rating Count","Maximum Installs","Good_App"], axis=1)
X.head()
```

```python
y = app_data_encoded["Good_App"].values
y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data
X_scaler = scaler.fit(X_train)

# Scaling the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating the decision tree classifier instance.
model = tree.DecisionTreeClassifier()
# Fitting the model.
model = model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = model.predict(X_test_scaled)
```

```python
predictions
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
acc_score
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test, y_pred)
```
